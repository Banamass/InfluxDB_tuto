```python
from influxdb_client import InfluxDBClient, Point
from influxdb_client.client.write_api import SYNCHRONOUS
import random
import time

client = InfluxDBClient(
    url="http://influxdb2:8086",
    token="MyInitialAdminToken0==",
    org="docs"
)

write_api = client.write_api(write_options=SYNCHRONOUS)

rooms = ["LivingRoom", "Kitchen"]

try:
    while True:
        temp = round(random.uniform(20, 28), 2)
        hum = round(random.uniform(38, 42), 2)
        room = random.choice(rooms)

        point = (
            Point("home")
            .tag("room", room)
            .field("temp", temp)
            .field("hum", hum)
        )

        write_api.write(bucket="home", org="docs", record=point)
        print(f"Sent -> room={room}, temp={temp}, hum={hum}")
        time.sleep(2)

except KeyboardInterrupt:
    print("Arrêt du script")

finally:
    client.close()
```
